```python
# Matplotlib Part - 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import random

# subplot()
# it allows us to create multiple plots within a single figure

# plot 1

x1 = random.randint(10,size=(7))
y1 = random.randint(6,size=(7))
plt.subplot(1,2,1)
plt.plot(x1,y1,color='green')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot 1')

# plot 2

x2 = random.randint(20,size=(7))
y2 = random.randint(10,size=(7))
plt.subplot(1,2,2)
plt.plot(x2,y2,color='blue')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot 2')

plt.show()
```

```python
x = random.randint(10,size=(7))
y = random.randint(6,size=(7))
color = ['cyan', 'blue', 'skyblue', 'pink', 'orange','red']

for i in range(1,7):
  plt.subplot(2,3,i)
  plt.plot(x,y,color=color[i-1])
  plt.title(f'Plot {i}')
  plt.xlabel('X axis')
  plt.ylabel('Y axis')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.7,
                    hspace=0.7)
plt.show()
```

```python
# Set up the 2x3 grid of subplots
plt.figure(figsize=(12, 6))  # Optional: Adjust figure size

# Define a list of colors for each plot
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan']

# Loop to generate 6 plots
for i in range(6):
    x = random.randint(100, size=7)
    y = random.randint(100, size=7)
    plt.subplot(2, 3, i + 1)  # 2 rows, 3 columns, subplot index (starts at 1)
    plt.plot(x, y, color=colors[i])
    plt.title(f'Plot {i+1}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

# Show all subplots in one figure
plt.tight_layout()  # Avoid overlap
plt.show()
```
